```python
#Mecab을 활용한 문장 분석 (품사 태깅)
from konlpy.tag import Mecab

mecab = Mecab()
print(mecab.pos('이 밤 그날의 반딧불을 당신의 창 가까이 보낼게요'))
```

```python
# 데이터 형식 맞추기
if 'percent' in positive_data.columns:
  positive_data = positive_data.drop(columns=['percent'])
positive_data_listed = positive_data.values.tolist()
print(positive_data_listed)
```

```python
# 데이터 형식 맞추기
if 'percent' in negative_data.columns:
  negative_data = negative_data.drop(columns=['percent'])
negative_data_listed = negative_data.values.tolist()
print(negative_data_listed)
```

```python
import json
import os
from tqdm import tqdm

def tokenize(doc):
    # norm은 정규화, stem은 근어로 표시하기를 나타냄
    return ['/'.join(t) for t in mecab.pos(doc)]

if os.path.isfile('/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/tokenize/positive_data_docs.json'):
    print('yy')
    with open('/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/tokenize/positive_data_docs.json') as f:
        positive_docs = json.load(f)
    with open('/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/tokenize/negative_data_docs.json') as f:
        negative_docs = json.load(f)
else:
    print('nn')
    positive_docs = []
    for row in tqdm(positive_data_listed):
        positive_docs.append((tokenize(row[0]), row[1]))

    negative_docs = []
    for row in tqdm(negative_data_listed):
        negative_docs.append((tokenize(row[0]), row[1]))

    # JSON 파일로 저장
    with open('/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/tokenize/positive_data_docs.json', 'w', encoding="utf-8") as make_file:
        json.dump(positive_docs, make_file, ensure_ascii=False, indent="\t")
    with open('/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/tokenize/negative_data_docs.json', 'w', encoding="utf-8") as make_file:
        json.dump(negative_docs, make_file, ensure_ascii=False, indent="\t")

print(positive_docs[0])
```

```python
for doc, label in positive_docs[:5]:
    print(doc, label)
```

```python
# 1음절 단어를 제거하고, 체언, 용언, 외국어만 남기고 나머지를 삭제하는 함수
def filter_tokens(doc):
    allowed_pos = {'NNG', 'NNP', 'NNB', 'NP', 'VV', 'VA', 'VX', 'SL'}
    filtered = []
    for token in doc:
        word, pos = token.split('/')
        if len(word) > 1 and pos in allowed_pos:
            filtered.append(token)
    return filtered

# positive 토큰 필터링
filtered_positive_docs = [(filter_tokens(doc), label) for doc, label in positive_docs]

# 결과 출력
for doc, label in filtered_positive_docs[:5]:
    print(doc, label)
```

```python
for doc, label in negative_docs[:5]:
    print(doc, label)
```

```python
# negative 토큰 필터링
filtered_negative_docs = [(filter_tokens(doc), label) for doc, label in negative_docs]

# 결과 출력
for doc, label in filtered_negative_docs[:5]:
    print(doc, label)
```

```python
# 단어 통일 함수
def unify_tokens(doc):
    unified = []
    for token in doc:
        word, pos = token.split('/')
        if word.lower() in {'ai', 'gpt'}:
            if word.lower() == 'ai':
                word = 'AI'
            elif word.lower() == 'gpt':
                word = 'gpt'
            elif word.lower() == 'it':
                word = 'it'
        unified.append(f'{word}/{pos}')
    return unified

# positive 문서 필터링
unified_positive_docs = [(unify_tokens(doc), label) for doc, label in filtered_positive_docs]

# 결과 출력
for doc, label in unified_positive_docs[:5]:
    print(doc, label)
```

```python
# negative 문서 필터링
unified_negative_docs = [(unify_tokens(doc), label) for doc, label in filtered_negative_docs]

# 결과 출력
for doc, label in unified_negative_docs[:5]:
    print(doc, label)
```

## 빈도 분석

```python
from collections import Counter

# 단어 빈도 계산을 위한 리스트 생성
all_positive_tokens = [token for doc, _ in unified_positive_docs for token in doc]

# 단어 빈도 계산
positive_word_freq = Counter(all_positive_tokens)

# unique 토큰 개수 출력
positive_unique_token_count = len(positive_word_freq)
print(f'Positive unique token count: {positive_unique_token_count}')

# 상위 100개 빈도의 단어 나열
positive_top_50_words = positive_word_freq.most_common(50)

# 결과 출력
for word, freq in positive_top_50_words:
    print(f'{word}: {freq}')
```

```python
# 빈도 데이터프레임으로 만들기
positive_top_50_df = pd.DataFrame(positive_top_50_words, columns=['단어', '빈도수'])
positive_top_50_df['단어'] = positive_top_50_df['단어'].apply(lambda x: x.split('/')[0])

positive_top_50_df
```

```python
# 단어 빈도 계산을 위한 리스트 생성
all_negative_tokens = [token for doc, _ in unified_negative_docs for token in doc]

# 단어 빈도 계산
negative_word_freq = Counter(all_negative_tokens)

# unique 토큰 개수 출력
negative_unique_token_count = len(negative_word_freq)
print(f'Negative unique token count: {negative_unique_token_count}')

# 상위 100개 빈도의 단어 나열
negative_top_50_words = negative_word_freq.most_common(50)

# 결과 출력
for word, freq in negative_top_50_words:
    print(f'{word}: {freq}')
```

```python
# 빈도 데이터프레임으로 만들기
negative_top_50_df = pd.DataFrame(negative_top_50_words, columns=['단어', '빈도수'])
negative_top_50_df['단어'] = negative_top_50_df['단어'].apply(lambda x: x.split('/')[0])

negative_top_50_df
```

```python
#csv 파일로 저장
positive_path_to_save = '/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/network/positive_fq50_freq.csv'
negative_path_to_save = '/content/drive/MyDrive/대학자료/대4/4학년 2학기/국어정보학/기말보고서/data/network/negative_fq50_freq.csv'
positive_top_50_df.to_csv(positive_path_to_save, index=True)
negative_top_50_df.to_csv(negative_path_to_save, index=True)
```

## 네트워크 분석

```python
import numpy as np
# 상위 50개 빈도의 토큰 리스트
positive_top_50_tokens = [word for word, _ in positive_top_50_words]

# 초기화된 50x50 행렬 생성
co_occurrence_matrix = np.zeros((50, 50), dtype=int)

# 토큰 인덱스 매핑 생성
token_index = {token: idx for idx, token in enumerate(positive_top_50_tokens)}

# 공동 출현 계산
for doc, _ in unified_positive_docs:
    # 상위 50개 토큰만 필터링
    tokens_in_doc = [token for token in doc if token in token_index]
    # 각 토큰 쌍의 공동 출현을 계산
    for i in range(len(tokens_in_doc)):
        for j in range(i + 1, len(tokens_in_doc)):
            token1, token2 = tokens_in_doc[i], tokens_in_doc[j]
            idx1, idx2 = token_index[token1], token_index[token2]
            co_occurrence_matrix[idx1, idx2] += 1
            co_occurrence_matrix[idx2, idx1] += 1  # 대칭 행렬

# 2mode 표를 데이터프레임으로 변환
positive_co_occurrence_df = pd.DataFrame(co_occurrence_matrix, index=positive_top_50_tokens, columns=positive_top_50_tokens)

# 결과 출력
positive_co_occurrence_df
```

```python
# POS 태그 제거 함수
def remove_pos_tagging(df):
    df.columns = [col.split('/')[0] for col in df.columns]
    df.index = [idx.split('/')[0] for idx in df.index]
    return df

# POS 태그 제거
positive_co_occurrence_removed_pos_df = remove_pos_tagging(positive_co_occurrence_df)

positive_co_occurrence_removed_pos_df
```

```python
# 상위 50개 빈도의 토큰 리스트
negative_top_50_tokens = [word for word, _ in negative_top_50_words]

# 초기화된 50x50 행렬 생성
co_occurrence_matrix = np.zeros((50, 50), dtype=int)

# 토큰 인덱스 매핑 생성
token_index = {token: idx for idx, token in enumerate(negative_top_50_tokens)}

# 공동 출현 계산
for doc, _ in unified_negative_docs:
    # 상위 50개 토큰만 필터링
    tokens_in_doc = [token for token in doc if token in token_index]
    # 각 토큰 쌍의 공동 출현을 계산
    for i in range(len(tokens_in_doc)):
        for j in range(i + 1, len(tokens_in_doc)):
            token1, token2 = tokens_in_doc[i], tokens_in_doc[j]
            idx1, idx2 = token_index[token1], token_index[token2]
            co_occurrence_matrix[idx1, idx2] += 1
            co_occurrence_matrix[idx2, idx1] += 1  # 대칭 행렬

# 2mode 표를 데이터프레임으로 변환
negative_co_occurrence_df = pd.DataFrame(co_occurrence_matrix, index=negative_top_50_tokens, columns=negative_top_50_tokens)

# 결과 출력
negative_co_occurrence_df
```

```python
negative_co_occurrence_removed_pos_df = remove_pos_tagging(negative_co_occurrence_df)

negative_co_occurrence_removed_pos_df
```
